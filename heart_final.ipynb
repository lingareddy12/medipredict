```python
import numpy as np

import pandas as pd
```

```python
data=pd.read_csv('cardio_train.csv',delimiter=";") 
data=data.abs()
data.head()
```

```python

data['gender'] = data['gender'].replace({1:0,2:1}).astype(np.uint8)   #female -0 male 1
data.shape
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data.describe()
```

```python
data["age"]=data["age"]/365.25
data=data.drop(columns=["id"])
data = data.drop(data[data['ap_lo'] <= 50].index) 
data = data.drop(data[data['ap_lo'] >=150].index) 
data = data.drop(data[data['ap_hi'] <= 50].index)     
data = data.drop(data[data['ap_hi'] >=250].index)
```

```python
data.describe()
```

```python
correlation_matrix = data.corr()

# Correlation of y with each feature column
correlation_with_y = correlation_matrix.iloc[:-1, -1]
correlation_with_y
```

```python
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd

# Assuming you have a DataFrame called 'data' containing your feature matrix X and target variable y

# Separate the feature matrix X and target variable y
X = data.drop('cardio', axis=1)  # Replace 'target_variable' with the actual column name of your target variable
y = data['cardio']

# Apply the chi-square test for feature selection
k = 8 # Specify the desired number of features to select
selector = SelectKBest(score_func=chi2, k=k)
X_new = selector.fit_transform(X, y)

# Get the selected feature names
selected_feature_names = X.columns[selector.get_support()].tolist()

# Create a new DataFrame with the selected features
X = pd.DataFrame(X_new, columns=selected_feature_names)

# Print the selected features
print("Selected Features:")
print(X.head())
```
